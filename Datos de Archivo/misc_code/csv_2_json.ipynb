```python
import pandas as pd
import json

'''
Para transformar de csv a json.

Lo empleo para los datos de NEA, que no permiten bajarlos en formato .json.

Se guardan en la misma carpeta donde está este archivo, big_data, y luego con la función
cargar_datos_NEA cojo todos los .jsons que estén en esta carpeta, y creo un data frame con ellos.
'''

# Nombre del archivo
input_csv = 'toi_4010_csv.csv'
output_json = 'toi_data_4010.json'

df = pd.read_csv(input_csv, comment='#')
json_data = df.to_json(orient='records', indent=4)

with open(output_json, 'w') as f:
    f.write(json_data)

print(df.head(2).to_json(orient='records', indent=4))
```
